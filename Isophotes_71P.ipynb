```python
import astropy.io.fits as pf
import numpy as np
import glob
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
%matplotlib inline

imloc = glob.glob("C:/Users/dirmu/Documents/fieldtrip/AllSpeed_71P/*.fits")

for name in imloc:
    O = pf.open(name)
    D = O[0].data

    size = len(D)
    x = np.arange(0,size,1)
    y = np.arange(0,size,1)
    X, Y = np.meshgrid(x,y)
    
#     Increase brigthness
#     There must be a better way of doing this
    D_b = 0
    D_b = D*1000

    plt.imshow(D,cmap='gray', norm=LogNorm())
    plt.contour(X,Y,D)
    plt.grid()
    plt.xlim(900,1050)
    plt.ylim(850,1000)
    plt.show()
```
